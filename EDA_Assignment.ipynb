**Count Plots**

```python
ax = sns.countplot(data['processor_brand'])
ax.set_title("Laptop counts")
plt.show()
# count the occurrences of each unique value in the 'processor_brand' column of the DataFrame data and visualizes it.
```

```python
ax = sns.countplot(data['processor_name'])
ax.set_title("brand")
plt.show()
#The count plot will show the frequency/count of each unique processor name in the dataset.
```

**BarPlot**

```python
data['processor_brand'].value_counts().index
```

```python
sns.barplot(x=data['processor_brand'].value_counts().index, y=data['processor_brand'].value_counts())
plt.show()
# It takes the unique processor brands as the x-axis values and their corresponding counts as the y-axis values.
```

```python
def barplot(x, y, hue):
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    sns.barplot(x=x, y=y, hue=hue, data=data, orient="h")

for y in ['brand', 'processor_brand']:
    barplot('warranty', y, 'brand')
    #it creates horizontal bar plots for the 'brand' and 'processor_brand' variables, both with 'warranty' on the x-axis, and colors the bars based on the 'brand' variable.
```

**Kde** **Plot**

```python
sns.histplot(data['Price'], bins=10,kde=True)
plt.show()
#This code will generate a histogram of the 'Price' column, showing the distribution of prices in the dataset,  with a KDE to visualize the probability density function of the data given
```

**Joint** **PLOT**

```python
sns.jointplot(x='brand', y='Price', data=data, kind='kde', space=1)
plt.show()
#This joint plot will display the relationship between the 'brand' and 'Price' variables using kde, providing insights into the distribution of prices for each brand.
```

**BOX PLOT**

```python
sns.boxplot(x='brand', y='Price', data=data)
plt.title('Boxplot of brand by price')
plt.show()
#
# the box plot illustrates the distribution of laptop prices for each brand, allowing for easy comparison of price distributions across different brands
```

**Violin Plot**

```python
plt.figure(figsize=(15, 15))
sns.violinplot(data=data, x="processor_brand", y="Price" )
plt.show()
#Violin plots are useful for visualizing the distribution of a continuous variable within different categories, this   plot illustrates the distribution of laptop prices for each processor brand, helping to compare the price distributions across different brands
```

```python
plt.figure(figsize=(10, 7))
sns.violinplot(x="brand", y="Price", hue="warranty", data=data, split=True)

plt.title("Violin Plot of Price by Brand and Warranty ")
plt.xlabel("Brand")
plt.ylabel("Price")

plt.show()
```

**Corelation** **HeatMap**

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
# in the DataFrame data, creates a heatmap to visualize the correlations between these variables using Seaborn.
```

**Pair Plot**

```python
sns.pairplot(data, hue='os')
plt.show()
#This pairplot will allow you to visually inspect how different variables correlate with each other across different operating system categories, as each subplot will be colored different  on basis of  operating system category
```

**Pie Chart**

```python
counts = data['processor_name'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
#it shows processor_name with their percentage present the dataset we used to plot this pie chart.
```

```python

counts = data['brand'].value_counts()

explode = [0]*len(counts)
explode[0] = 0.1

plt.figure(figsize=(10, 10))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', explode=explode)
plt.title('Pie Chart')
plt.show()
#this pie chart shows brand contribution with their percentage and explodes (or highlights a particular value).
```
